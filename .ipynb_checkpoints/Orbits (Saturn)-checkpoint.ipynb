```python
import numpy as np
import matplotlib.pyplot as plt
from skyfield.api import load
from scipy.optimize import curve_fit
from matplotlib.pyplot import figure, show, subplots, tight_layout


data   = load('de421.bsp')
ts     = load.timescale()
t      = ts.utc(2023, 1, range(366 * 5))

print(t)

jupiter, earth  = data['Saturn barycenter'], data['Earth']
jpos, epos      = jupiter.at(t).position.au, earth.at(t).position.au
dis = (((jpos[0]-epos[0])**2+(jpos[1]-epos[1])**2+(jpos[2]-epos[2])**2)**0.5)


plt.plot(t.J-2023, dis, "bo", ms= 1)
plt.show()
```

```python
b = 6.00052
def fit(x, a, c, d, e, f):
    res = a*(np.sin(b/20.996*x+f)+e*np.sin(b*x+d))+c
    return res

a = 5
c = 5.3
d = 1
e = 5
f = 1



p0 = (a, c, d, e, f)

popt, pcov = curve_fit(fit, t.J-2023, dis, p0=p0)

print(popt)

fig = figure()
fig.set_size_inches(8, 8)
frame = fig.add_subplot(1,1,1)
frame.plot(t.J-2023, dis, "b")
frame.plot(t.J-2023, fit(t.J, *popt), "r-")
fig.set_size_inches(10, 5)
show()
```

```python
b=6.000

a = 0.40727115
c = 9.4680323
d = 1.10990923
e = 2.42241395
f = 2.05124696

x1 = np.linspace(0,np.max(t.J-2023),len(t))
y1 = [a*(np.sin(b/21*x1+f)+e*np.sin(b*x1+d))+c for x1 in x1]

fig = figure()
fig.set_size_inches(8, 8)
frame = fig.add_subplot(1,1,1)
frame.plot(t.J-2023, dis, "b")
frame.plot(t.J-2023, y1, "r")
fig.set_size_inches(10, 5)
show()
```

```python
from scipy.fft import fft, fftfreq

import numpy as np

# Number of sample points

N = 600

# sample spacing

T = 1.0 / 800.0

x = t.J-2023

y = dis

yf = fft(y)

xf = fftfreq(N, T)[:N//2]

import matplotlib.pyplot as plt

plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]), "o")
plt.ylim(0,5)
plt.xlim(0.01,50)
plt.xticks(np.arange(min(xf), 50, 1.0))
plt.grid()

plt.show()
```
