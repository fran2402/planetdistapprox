```python
import numpy as np
import matplotlib.pyplot as plt
from skyfield.api import load
from scipy.optimize import curve_fit
from matplotlib.pyplot import figure, show, subplots, tight_layout


data   = load('de421.bsp')
ts     = load.timescale()
t      = ts.utc(2023, 1, range(366 * 10))

print(t)

jupiter, earth  = data['Jupiter barycenter'], data['Earth']
jpos, epos      = jupiter.at(t).position.au, earth.at(t).position.au
dis = (((jpos[0]-epos[0])**2+(jpos[1]-epos[1])**2+(jpos[2]-epos[2])**2)**0.5)


plt.plot(t.J-2023, dis, "bo", ms= 1)
plt.show()
```

```python
b = 5.70551
def fit(x, a, c, d, e, f):
    res = a*(np.sin(b/11.203*x+f)+e*np.sin(b*x+d))+c
    return res

a = 5
c = 5.3
d = 1
e = 5
f = 1



p0 = (a, c, d, e, f)

popt, pcov = curve_fit(fit, t.J-2023, dis, p0=p0)

print(popt)

fig = figure()
fig.set_size_inches(8, 8)
frame = fig.add_subplot(1,1,1)
frame.plot(t.J-2023, dis, "b")
frame.plot(t.J-2023, fit(t.J, *popt), "r-")
fig.set_size_inches(10, 5)
show()
```
